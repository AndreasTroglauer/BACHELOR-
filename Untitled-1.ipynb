```python
import numpy as np
import matplotlib.pyplot as plt

# Define the transition matrix P (replace this with your actual transition matrix)
P = np.array([[0.507, 0.125],
              [0.492, 0.874],])

# Compute eigenvalues
eigenvalues = np.linalg.eigvals(P)

# Create a plot with square aspect ratio
plt.figure(figsize=(6,6))

# Plot eigenvalues
plt.scatter(eigenvalues.real, eigenvalues.imag, color='red')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Eigenvalues')
plt.xlabel('Real')
plt.ylabel('Imaginary')

# Set plot limits to [-1, 1] on both axes
plt.xlim([-1, 1])
plt.ylim([-1, 1])

# Add circle
circle = plt.Circle((0, 0), 1, color='blue', fill=False)
plt.gca().add_patch(circle)

# Show plot
plt.grid(True)
plt.show()
```
